```python
def convertNumToDatestr(num):
    year = num//10_000
    month= (num-year*10_000)//100
    day = (num-year*10_000-month*100)

    year_str = str(year).zfill(4)
    month_str = str(month).zfill(2)
    day_str = str(day).zfill(2)
    
    datestr = f"{year_str}-{month_str}-{day_str}"
    return pd.to_datetime(datestr)
```

```python
def style_table(df):

    def highlight_frac_pass(val):
        if val > 0.90:
            return "background-color: #b6f2b6"
        elif val < 0.80:
            return "background-color: #f5b5b5"
        return ""

    def highlight_frac_out(val):
        if val > 0.2:
            return "background-color: #f5b5b5"
        return ""

    return (
        df.style
        .applymap(highlight_frac_pass, subset=["frac_pass_all"])
        .applymap(highlight_frac_out, subset=["frac_out"])
    )
```
