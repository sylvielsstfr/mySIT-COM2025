```python
def style_table(df):

    def highlight_meanmedian(val):
        if np.abs(val) > 0.1:
            return "background-color: #f5b5b5"
        elif np.abs(val) < 0.05:
            return "background-color: #b6f2b6"
        return ""

    def highlight_width(val):
        if np.abs(val) > 0.2:
            return "background-color: #f5b5b5"
        elif np.abs(val) < 0.1:
            return "background-color: #b6f2b6"
        return ""

    return (
        df.style
        .applymap(highlight_meanmedian, subset=["mean","median"])
        .applymap(highlight_width, subset=["sigma","sigma_mad","sigma_iqr"])
    )
```
