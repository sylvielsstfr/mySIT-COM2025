## Style to plot tables

```python
def style_table(df):

    def highlight_meanmedian(val):
        if np.abs(val) > 0.1:
            return "background-color: #f5b5b5"
        elif np.abs(val) < 0.05:
            return "background-color: #b6f2b6"
        return ""

    def highlight_widthcore(val):
        if np.abs(val) > 0.2:
            return "background-color: #f5b5b5"
        elif np.abs(val) < 0.15:
            return "background-color: #b6f2b6"
        return ""

    def highlight_widthtail(val):
        if np.abs(val) > 0.3:
            return "background-color: #f5b5b5"
        elif np.abs(val) < 0.2:
            return "background-color: #b6f2b6"
        return ""

    return (
        df.style
        .applymap(highlight_meanmedian, subset=["mean","median"])
        .applymap(highlight_widthcore, subset=["sigma_mad","sigma_iqr"])
        .applymap(highlight_widthtail, subset=["sigma"])
    )
```

## Loop on files

```python
for index, df in enumerate(all_df):

    tag = all_tag[index]

    pdf_file = f"{pathdata}/{prefix}_{version_run}_pwv_performances_summary_{tag}.pdf"
    html_file = f"{pathdata}/{prefix}_{version_run}_pwv_performances_summary_{tag}.html"

    #print(pdf_file, html_file)

    caption = (
        f"PWV performances summary — {version_run} — {tag}"
    )

    with pd.option_context("display.max_rows", None,
                           "display.max_columns", None):

        styled = (
            style_table(df)
            .set_caption(caption)
            .format("{:.3f}")
            .set_table_styles([
                {'selector': 'caption',
                 'props': [('font-size', '16px'),
                           ('font-weight', 'bold'),
                           ('text-align', 'center')]}
            ])
        )

        display(styled)

        html_string = styled.to_html()
        styled.to_html(html_file)

        await html_string_to_pdf(html_string, pdf_file)
```
