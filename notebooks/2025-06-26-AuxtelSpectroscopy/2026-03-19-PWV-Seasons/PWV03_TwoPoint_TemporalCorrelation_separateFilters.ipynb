# PWV03 : Two-Point Temporal Correlation Function — per filter

## Overview

This notebook quantifies the **temporal coherence** of the precipitable water
vapour (PWV) measured by AuxTel/Spectractor at Cerro Pachon on intra-night
timescales ranging from one minute to ten hours.  Understanding this coherence
is essential for two reasons.  First, it sets the time window over which a
single PWV measurement can be considered representative of the atmospheric
conditions during a nearby exposure — a key input for photometric calibration
strategies.  Second, it characterises the dominant source of stochastic
atmospheric noise for wide-field surveys such as LSST/Rubin.

## Statistical estimator

The **two-point temporal correlation function** measures how strongly the PWV
fluctuation at time $t$ is correlated with the fluctuation at time $t + \Delta t$:
$$
C(\Delta t) =
  \frac{\bigl\langle \delta\mathrm{PWV}(t)\,\delta\mathrm{PWV}(t+\Delta t)\bigr\rangle}
       {\bigl\langle \delta\mathrm{PWV}^2 \bigr\rangle}
\qquad\text{with}\quad
\delta\mathrm{PWV}(t) = \mathrm{PWV}(t) - \langle\mathrm{PWV}\rangle_{\mathrm{night}(t)}.
$$
By construction $C(0) = 1$ (perfect self-correlation) and
$C(\Delta t) \to 0$ when the two measurements are statistically independent.
Negative values indicate an anti-correlation, which can arise from
quasi-periodic atmospheric structures passing over the telescope.

## Why night-by-night mean subtraction?

The absolute PWV level at Cerro Pachon typically varies by several millimetres
from one night to the next, driven by synoptic-scale weather systems and the
seasonal cycle.  Within a single observing night, however, the PWV is far more
stable and its fluctuations have a much smaller amplitude.  If a global mean
computed over the entire multi-year dataset were subtracted, the large
inter-night offsets would dominate the variance $\langle\delta\mathrm{PWV}^2\rangle$
and inflate the denominator of $C(\Delta t)$, artificially suppressing the
correlation at short lags.  Instead, we subtract the **nightly mean**
$\langle\mathrm{PWV}\rangle_{\mathrm{night}}$ computed separately for each
observing night (`nightObs`), so that $\delta\mathrm{PWV}$ reflects only the
genuine intra-night atmospheric fluctuations.

## Why pair-based estimation on an irregular grid?

AuxTel observations are **irregularly sampled** in time: the cadence within a
night varies with the observing programme, targets change, and gaps occur
between exposures.  A standard fast Fourier transform (FFT) requires uniform
sampling and would require interpolation, introducing artefacts.  Instead, we
use a direct **pair-based estimator**: all ordered pairs $(i, j)$ with
$j > i$ within the same night are formed, their time separation
$\delta t_{ij} = t_j - t_i$ and product
$\delta\mathrm{PWV}_i \cdot \delta\mathrm{PWV}_j$ are computed, and the pairs
are then averaged in logarithmically-spaced lag bins:
$$
C(\Delta t_k) =
  \frac{1}{N_k\,\sigma^2}
  \sum_{\{i,j\}\in\mathrm{bin}_k} \delta\mathrm{PWV}_i\,\delta\mathrm{PWV}_j,
$$
where $\sigma^2 = \langle\delta\mathrm{PWV}^2\rangle$ is the pooled
intra-night variance and $N_k$ is the pair count in bin $k$.  Logarithmic bins
give equal statistical weight to each decade of lag, which is appropriate
given that the PWV coherence is expected to span several orders of magnitude
in time.

## Why an exponential fit?

The simplest physical model for the decorrelation of a turbulent field advected
by a steady wind is an exponential decay:
$$
C(\Delta t) = A\,\exp\!\left(-\frac{\Delta t}{\tau}\right),
$$
where $\tau$ is the **decorrelation timescale** and $A \leq 1$ is an amplitude
that accounts for any white-noise floor at $\Delta t \to 0$ (e.g. measurement
errors, sub-minute atmospheric turbulence unresolved by the cadence).  Fitting
this model to the binned $C(\Delta t)$ serves several purposes:

1. **Compact summary.** The single parameter $\tau$ encapsulates the temporal
   coherence of the PWV in a form that can be quoted in a paper and compared
   with atmospheric models or other sites.
2. **Objective extraction.** The empirical 1/e crossing of the measured $C$
   depends on the bin width and the noise level.  A weighted least-squares fit
   provides a more robust and reproducible estimate with a propagated
   uncertainty.
3. **Physical interpretation.** If the Taylor frozen-turbulence hypothesis
   holds (i.e. the PWV field is carried rigidly by the wind), the decorrelation
   timescale $\tau \approx L / v$ where $L$ is the spatial coherence length
   and $v$ is the wind speed.  A measured $\tau$ can therefore be converted
   into a spatial coherence scale once the wind speed is known.
4. **Per-filter consistency check.** Fitting the same model to the correlation
   functions measured through the `empty`, `OG550`, and `FELH0600` filters
   allows us to verify that the decorrelation timescale is filter-independent
   — as expected for a purely atmospheric quantity — and to identify any
   systematic biases introduced by the different spectral bandpasses or
   observing strategies.

Two complementary representations are shown throughout:
- **Linear x-axis in hours** (0 – 10 h): gives an intuitive view of the
  decorrelation on the scale of an observing night.
- **Logarithmic x-axis in minutes** (1 – 600 min): compresses the full dynamic
  range and reveals any power-law-like behaviour at short lags.

## 3. Prepare PWV time series — night-by-night mean subtraction

### Physical motivation

The PWV at Cerro Pachon exhibits variability on two very different timescales.
On **inter-night** timescales (days to months), the mean level can shift by
several millimetres, driven by synoptic-scale meteorological systems, the
seasonal cycle of the ITCZ, and episodic intrusions of moist air from the
Pacific.  On **intra-night** timescales (minutes to hours), these large-scale
offsets are essentially frozen, and the PWV fluctuates around a quasi-constant
nightly mean with a much smaller amplitude, typically a fraction of a
millimetre to ~1 mm.

### Implementation

Subtracting a single global mean computed over the entire multi-year dataset
would conflate these two very different regimes.  The inter-night variance
would dominate $\sigma^2 = \langle\delta\mathrm{PWV}^2\rangle$, inflating the
denominator of $C(\Delta t)$ and suppressing the measured correlation at all
lags.  In the worst case, the correlation function would appear to be nearly
flat and close to zero even at short lags, giving the misleading impression
that the PWV is uncorrelated from one exposure to the next — which is
physically incorrect.

We therefore subtract the **nightly mean** $\langle\mathrm{PWV}\rangle_n$
computed independently for each observing night, identified by its `nightObs`
key (re-computed from the UTC timestamp after a Chile local-time offset of
$-12\,\mathrm{h}$ to avoid splitting a night across two calendar dates):
$$
\delta\mathrm{PWV}(t) = \mathrm{PWV}(t)
  - \langle\mathrm{PWV}\rangle_{\mathrm{night}(t)}.
$$
This is implemented efficiently in pandas as a `groupby + transform` operation
with no explicit loop.  The resulting residuals $\delta\mathrm{PWV}$ represent
the genuine atmospheric fluctuations within each night, and their pooled
variance $\sigma^2 = \langle\delta\mathrm{PWV}^2\rangle$ (computed over all
nights) serves as the normalisation constant for the correlation function.

```python
# ── Best-estimate PWV ─────────────────────────────────────────────────────────
df_keep["PWV"]     = df_keep["PWV [mm]_ram"]
df_keep["PWV_err"] = df_keep["PWV [mm]_err_ram"]

# ── Time in minutes since first observation ───────────────────────────────────
t0_min = df_keep["MJD"].min() * 1440.0
df_keep["t_min"] = df_keep["MJD"] * 1440.0 - t0_min

# ── Night-by-night mean subtraction ──────────────────────────────────────────
# Compute the mean PWV for each nightObs separately, then broadcast back.
night_means = df_keep.groupby("nightObs")["PWV"].transform("mean")
df_keep["PWV_night_mean"] = night_means
df_keep["dPWV"]           = df_keep["PWV"] - night_means

# Pooled intra-night variance (used as C(0) denominator for all pairs)
pwv_var = df_keep["dPWV"].var(ddof=1)

# ── Diagnostics ───────────────────────────────────────────────────────────────
n_nights = df_keep["nightObs"].nunique()
night_stats = df_keep.groupby("nightObs").agg(
    n_obs     = ("PWV", "count"),
    pwv_mean  = ("PWV", "mean"),
    pwv_std   = ("PWV", "std"),
    dpwv_std  = ("dPWV", "std"),
)

print(f"N observations              : {len(df_keep)}")
print(f"N nights (nightObs)         : {n_nights}")
print(f"Median obs per night        : {night_stats['n_obs'].median():.0f}")
print()
print(f"Night-mean PWV  — range     : {night_stats['pwv_mean'].min():.2f} – "
      f"{night_stats['pwv_mean'].max():.2f} mm")
print(f"Night-mean PWV  — std across nights : "
      f"{night_stats['pwv_mean'].std():.3f} mm  (inter-night scatter)")
print()
print(f"Intra-night std — median per night  : "
      f"{night_stats['dpwv_std'].median():.3f} mm")
print(f"Pooled intra-night variance  sigma^2 : {pwv_var:.4f} mm^2  "
      f"=> sigma = {np.sqrt(pwv_var):.3f} mm")
print(f"Time span                   : {df_keep['t_min'].max()/1440:.1f} days  "
      f"({df_keep['t_min'].max()/60:.0f} hours)")

display(night_stats.describe().round(3))
```

```python
# ── Quick sanity plot: inter-night vs intra-night variability ─────────────────
fig0, axes0 = plt.subplots(1, 2, figsize=(14, 4))

# Left: nightly mean PWV over time
night_order = night_stats.sort_index()
axes0[0].errorbar(
    range(len(night_order)),
    night_order["pwv_mean"],
    yerr=night_order["pwv_std"],
    fmt="o", color="steelblue", ms=4, elinewidth=0.8, capsize=2,
    label="Nightly mean +/- std"
)
axes0[0].set_xlabel("Night index (chronological)", fontsize=12)
axes0[0].set_ylabel("PWV [mm]", fontsize=12)
axes0[0].set_title("Nightly mean PWV — inter-night variability", fontsize=11)
axes0[0].legend(fontsize=9)

# Right: histogram of dPWV (intra-night residuals)
axes0[1].hist(df_keep["dPWV"].dropna(), bins=60, color="steelblue",
              alpha=0.8, edgecolor="white", linewidth=0.3)
axes0[1].axvline(0, color="black", ls="--", lw=1)
axes0[1].set_xlabel(r"$\delta\mathrm{PWV}$ [mm]", fontsize=12)
axes0[1].set_ylabel("Count", fontsize=12)
axes0[1].set_title(
    rf"Intra-night residuals  $\sigma = {np.sqrt(pwv_var):.3f}$ mm",
    fontsize=11
)

fig0.suptitle(
    f"Night-by-night mean subtraction — {version_run}",
    fontsize=12, y=1.02
)
fig0.tight_layout()
figfile0 = f"{pathfigs}/{prefix}_{version_run}_night_mean_subtraction{figtype}"
fig0.savefig(figfile0, bbox_inches="tight")
print(f"Saved: {figfile0}")
plt.show()
```

## 4. Two-point correlation — binned pair estimator

All measurement pairs $(i, j)$ with $j > i$ **belonging to the same night**
are formed.  The same-night constraint is enforced explicitly by looping over
each `nightObs` group, which guarantees that no inter-night pair can
accidentally fall within the 10-hour lag window (which could happen if two
consecutive nights overlap near midnight in local time).  The lag
$\delta t_{ij} = t_j - t_i$ is expressed in minutes and retained only if it
falls within $[1, 600]\,\mathrm{min}$.  The products and lags from all nights
are then concatenated and binned in 50 logarithmically-spaced bins covering
this range.

```python
# ── Parameters ────────────────────────────────────────────────────────────────
LAG_MIN_MIN     =   1.0    # minimum lag: 1 minute
LAG_MAX_MIN     = 600.0    # maximum lag: 10 hours = 600 minutes
N_BINS_LOG      =  50      # number of log-spaced bins
N_MAX_SUBSAMPLE = 3000     # max points per night for pair computation

rng = np.random.default_rng(seed=42)

nights = sorted(df_keep["nightObs"].unique())
print(f"Processing {len(nights)} nights...")
```

```python
# ── Compute pairs night-by-night (guarantees same-night constraint) ────────────
# Accumulate (dt, product) arrays across all nights, then bin once.

dt_pairs_list   = []
prod_pairs_list = []

for night in nights:
    sub_n = df_keep[df_keep["nightObs"] == night]
    if len(sub_n) < 2:
        continue

    t_n  = sub_n["t_min"].values
    dp_n = sub_n["dPWV"].values

    # optional sub-sampling per night (unlikely needed, but safe)
    if len(t_n) > N_MAX_SUBSAMPLE:
        idx_n = np.sort(rng.choice(len(t_n), N_MAX_SUBSAMPLE, replace=False))
        t_n   = t_n[idx_n]
        dp_n  = dp_n[idx_n]

    i_n, j_n = np.triu_indices(len(t_n), k=1)
    dt_n     = t_n[j_n] - t_n[i_n]
    prod_n   = dp_n[i_n] * dp_n[j_n]

    # keep only pairs within the 10-h window
    ok_n = (dt_n >= LAG_MIN_MIN) & (dt_n <= LAG_MAX_MIN)
    dt_pairs_list.append(dt_n[ok_n])
    prod_pairs_list.append(prod_n[ok_n])

dt_pairs   = np.concatenate(dt_pairs_list)
prod_pairs = np.concatenate(prod_pairs_list)

print(f"Total same-night pairs within 10 h : {len(dt_pairs):,}")
if len(dt_pairs) > 0:
    print(f"Lag range (kept) : {dt_pairs.min():.2f} – {dt_pairs.max():.1f} min")
```

```python
# ── Log-spaced lag bins over [1 min, 600 min] ─────────────────────────────────
bin_edges = np.logspace(
    np.log10(LAG_MIN_MIN),
    np.log10(LAG_MAX_MIN),
    N_BINS_LOG + 1
)
bin_centers_min = np.sqrt(bin_edges[:-1] * bin_edges[1:])  # geometric mean

bin_idx = np.digitize(dt_pairs, bin_edges) - 1
valid   = (bin_idx >= 0) & (bin_idx < N_BINS_LOG)

corr     = np.full(N_BINS_LOG, np.nan)
n_pairs  = np.zeros(N_BINS_LOG, dtype=int)
corr_err = np.full(N_BINS_LOG, np.nan)

for k in range(N_BINS_LOG):
    mask = valid & (bin_idx == k)
    nk   = mask.sum()
    n_pairs[k] = nk
    if nk == 0:
        continue
    prods_k     = prod_pairs[mask]
    corr[k]     = prods_k.mean() / pwv_var
    corr_err[k] = prods_k.std(ddof=1) / (np.sqrt(nk) * pwv_var)

lag_min = bin_centers_min
lag_hr  = bin_centers_min / 60.0

MIN_PAIRS = 5
good = n_pairs >= MIN_PAIRS
print(f"Bins with >= {MIN_PAIRS} pairs : {good.sum()} / {N_BINS_LOG}")
```

---
## 5. Decorrelation timescale

```python
df_corr = pd.DataFrame({
    "lag_min" : lag_min[good],
    "lag_hr"  : lag_hr[good],
    "C"       : corr[good],
    "C_err"   : corr_err[good],
    "n_pairs" : n_pairs[good],
})

inv_e = 1.0 / np.e
cross = df_corr[df_corr["C"] < inv_e]
if len(cross) > 0:
    tau_min = cross.iloc[0]["lag_min"]
    tau_hr  = cross.iloc[0]["lag_hr"]
    print(f"Decorrelation timescale (C < 1/e) : {tau_min:.1f} min = {tau_hr:.2f} h")
else:
    tau_min = np.nan
    tau_hr  = np.nan
    print("C(Delta_t) does not drop below 1/e within 10 h")

display(df_corr)
```

## 6. Main figure — 2-panel  (linear hours | log minutes)

Left panel : x-axis = lag in **hours**, **linear** scale, range 0 – 10 h  
Right panel : x-axis = lag in **minutes**, **log** scale, range 1 – 600 min  
Both y-axes : $C(\Delta t)$, linear scale.

```python
CORR_COLOR  = "steelblue"
ERR_COLOR   = "steelblue"
ZERO_COLOR  = "gray"
INV_E_COLOR = "darkorange"
TAU_COLOR   = "darkred"


def _decorate_corr_ax(ax, x_tau=None, xlabel="", xscale="linear",
                      xlim=None, ylim=(-0.5, 1.5)):
    ax.axhline(0.0,      ls="--", color=ZERO_COLOR,  lw=1.0, zorder=1)
    ax.axhline(1.0/np.e, ls=":",  color=INV_E_COLOR, lw=1.5, zorder=1,
               label=r"$C = 1/e$")
    if x_tau is not None and np.isfinite(x_tau):
        unit = xlabel.split('[')[1].rstrip(']') if '[' in xlabel else ''
        ax.axvline(x_tau, ls="-", color=TAU_COLOR, lw=1.5, zorder=2,
                   label=rf"$\tau_{{1/e}} = {x_tau:.1f}$ {unit}")
    ax.set_ylabel(r"$C(\Delta t)$", fontsize=13)
    ax.set_xlabel(xlabel, fontsize=13)
    ax.set_xscale(xscale)
    if xlim is not None:
        ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.legend(fontsize=9, loc="upper right")


fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(wspace=0.32)

# ── LEFT : linear x in hours, 0 – 10 h ────────────────────────────────────────
ax_lin = axes[0]
ax_lin.fill_between(
    lag_hr[good], corr[good] - corr_err[good], corr[good] + corr_err[good],
    alpha=0.25, color=ERR_COLOR
)
ax_lin.plot(lag_hr[good], corr[good],
            color=CORR_COLOR, lw=1.8, marker="o", ms=3.5, zorder=3,
            label=r"$C(\Delta t)$")
_decorate_corr_ax(
    ax_lin,
    x_tau  = tau_hr if np.isfinite(tau_hr) else None,
    xlabel = r"$\Delta t$ [hours]",
    xscale = "linear",
    xlim   = (0, LAG_MAX_MIN / 60.0),
)
ax_lin.set_title("Two-point correlation — linear hours", fontsize=11)

# ── RIGHT : log x in minutes, 1 – 600 min ─────────────────────────────────────
ax_log = axes[1]
ax_log.fill_between(
    lag_min[good], corr[good] - corr_err[good], corr[good] + corr_err[good],
    alpha=0.25, color=ERR_COLOR
)
ax_log.plot(lag_min[good], corr[good],
            color=CORR_COLOR, lw=1.8, marker="o", ms=3.5, zorder=3,
            label=r"$C(\Delta t)$")
_decorate_corr_ax(
    ax_log,
    x_tau  = tau_min if np.isfinite(tau_min) else None,
    xlabel = r"$\Delta t$ [minutes]",
    xscale = "log",
    xlim   = (LAG_MIN_MIN, LAG_MAX_MIN),
)
ax_log.set_title("Two-point correlation — log minutes", fontsize=11)

# secondary top axis: hours on the log panel
ax_log2 = ax_log.twiny()
ax_log2.set_xlim(np.array(ax_log.get_xlim()) / 60.0)
ax_log2.set_xscale("log")
ax_log2.set_xlabel(r"$\Delta t$ [hours]", fontsize=10, labelpad=6)

fig.suptitle(
    f"PWV two-point correlation (same-night pairs, 0–10 h)\n"
    f"night-by-night mean subtracted — {version_run} (tight cuts)",
    fontsize=11, y=1.02
)

figfile1 = f"{pathfigs}/{prefix}_{version_run}_2point_corr_2panel{figtype}"
fig.savefig(figfile1, bbox_inches="tight")
print(f"Saved: {figfile1}")
plt.show()
```

## 9. Separate correlations per filter

```python
FILTERS_OF_INTEREST = ["empty", "OG550_65mm_1", "FELH0600"]
filter_color = {
    "empty"        : "#1f77b4",
    "OG550_65mm_1" : "#d62728",
    "FELH0600"     : "#2ca02c",
}
filter_label = {
    "empty"        : "empty",
    "OG550_65mm_1" : "OG550",
    "FELH0600"     : "FELH0600",
}

corr_by_filter     = {}
corr_err_by_filter = {}
npairs_by_filter   = {}

for filt in FILTERS_OF_INTEREST:
    sub_f = df_keep[df_keep["FILTER"] == filt]
    if len(sub_f) < 10:
        print(f"  {filt}: too few points ({len(sub_f)}), skipped")
        continue

    # Use the filter-specific nightly mean subtraction
    # (dPWV was already computed on the full df_keep with all filters;
    # here we recompute it restricted to this filter so that the
    # nightly mean reflects only observations through that filter)
    sub_f = sub_f.copy()
    night_means_f = sub_f.groupby("nightObs")["PWV"].transform("mean")
    sub_f["dPWV_f"] = sub_f["PWV"] - night_means_f
    var_f = sub_f["dPWV_f"].var(ddof=1)
    if var_f == 0 or not np.isfinite(var_f):
        continue

    dt_list_f, prod_list_f = [], []
    for night in sub_f["nightObs"].unique():
        sub_fn = sub_f[sub_f["nightObs"] == night]
        if len(sub_fn) < 2:
            continue
        t_fn  = sub_fn["t_min"].values
        dp_fn = sub_fn["dPWV_f"].values
        if len(t_fn) > N_MAX_SUBSAMPLE:
            idx_fn = np.sort(rng.choice(len(t_fn), N_MAX_SUBSAMPLE, replace=False))
            t_fn   = t_fn[idx_fn]
            dp_fn  = dp_fn[idx_fn]
        i_fn, j_fn = np.triu_indices(len(t_fn), k=1)
        dt_fn      = t_fn[j_fn] - t_fn[i_fn]
        prod_fn    = dp_fn[i_fn] * dp_fn[j_fn]
        ok_fn      = (dt_fn >= LAG_MIN_MIN) & (dt_fn <= LAG_MAX_MIN)
        dt_list_f.append(dt_fn[ok_fn])
        prod_list_f.append(prod_fn[ok_fn])

    if not dt_list_f:
        continue
    dt_f   = np.concatenate(dt_list_f)
    prod_f = np.concatenate(prod_list_f)

    bidx_f = np.digitize(dt_f, bin_edges) - 1
    val_f  = (bidx_f >= 0) & (bidx_f < N_BINS_LOG)

    c_f  = np.full(N_BINS_LOG, np.nan)
    ce_f = np.full(N_BINS_LOG, np.nan)
    np_f = np.zeros(N_BINS_LOG, dtype=int)

    for k in range(N_BINS_LOG):
        m  = val_f & (bidx_f == k)
        nk = m.sum()
        np_f[k] = nk
        if nk < MIN_PAIRS:
            continue
        pk = prod_f[m]
        c_f[k]  = pk.mean() / var_f
        ce_f[k] = pk.std(ddof=1) / (np.sqrt(nk) * var_f)

    corr_by_filter[filt]     = c_f
    corr_err_by_filter[filt] = ce_f
    npairs_by_filter[filt]   = np_f
    print(f"  {filt}: {len(sub_f)} obs, {len(dt_f):,} same-night pairs")

print("Done.")
```

```python
fig4, axes4 = plt.subplots(1, 2, figsize=(16, 6))
fig4.subplots_adjust(wspace=0.32)

for ax4, xdata, xlabel, xscale, xlim in [
    (axes4[0], lag_hr,  r"$\Delta t$ [hours]",  "linear",
     (0, LAG_MAX_MIN / 60.0)),
    (axes4[1], lag_min, r"$\Delta t$ [minutes]", "log",
     (LAG_MIN_MIN, LAG_MAX_MIN)),
]:
    ax4.plot(xdata[good], corr[good],
             color="black", lw=1.0, alpha=0.4, ls="--", label="All filters")
    for filt in FILTERS_OF_INTEREST:
        if filt not in corr_by_filter:
            continue
        c_f  = corr_by_filter[filt]
        ce_f = corr_err_by_filter[filt]
        gf   = npairs_by_filter[filt] >= MIN_PAIRS
        ax4.fill_between(
            xdata[gf], c_f[gf] - ce_f[gf], c_f[gf] + ce_f[gf],
            alpha=0.15, color=filter_color[filt]
        )
        ax4.plot(xdata[gf], c_f[gf],
                 color=filter_color[filt], lw=1.8, marker="o", ms=3,
                 label=filter_label[filt])
    ax4.axhline(0.0,      ls="--", color=ZERO_COLOR,  lw=1.0)
    ax4.axhline(1.0/np.e, ls=":",  color=INV_E_COLOR, lw=1.5,
                label=r"$C = 1/e$")
    ax4.set_xscale(xscale)
    ax4.set_xlim(xlim)
    ax4.set_ylim(-1.35, 1.35)
    ax4.set_xlabel(xlabel, fontsize=13)
    ax4.set_ylabel(r"$C(\Delta t)$", fontsize=13)
    ax4.legend(fontsize=9, loc="upper right")

axes4[0].set_title("Per-filter correlation — linear hours", fontsize=11)
axes4[1].set_title("Per-filter correlation — log minutes",  fontsize=11)
fig4.suptitle(
    f"PWV two-point correlation by filter (0–10 h, same-night) — {version_run}",
    fontsize=12, y=1.01
)
figfile4 = f"{pathfigs}/{prefix}_{version_run}_2point_corr_per_filter{figtype}"
fig4.savefig(figfile4, bbox_inches="tight")
print(f"Saved: {figfile4}")
plt.show()
```

## 10. Exponential fit of the decorrelation timescale

### Model and motivation

We fit the measured $C(\Delta t)$ with an exponential decay model:
$$
C_\mathrm{model}(\Delta t) = A\,\exp\!\left(-\frac{\Delta t}{\tau}\right),
$$
where $\tau > 0$ is the **decorrelation timescale** (in minutes) and
$A \in (0, 1]$ is an overall amplitude.  The amplitude $A$ being slightly
below unity signals the presence of a noise floor: at the shortest accessible
lags (a few minutes), consecutive exposures already differ by a small
uncorrelated component arising from measurement uncertainties and sub-minute
atmospheric turbulence not resolved by the observing cadence.

The exponential form is motivated by the **Taylor frozen-turbulence**
hypothesis, which states that the turbulent field (here, the PWV column) is
advected rigidly by the large-scale wind without evolving intrinsically on the
timescale of the observations.  Under this assumption, the temporal correlation
function and the spatial correlation function are related by
$C(\Delta t) = C_\mathrm{spatial}(v\,\Delta t)$, where $v$ is the wind speed.
If the spatial correlation is itself exponential with coherence length $L$,
one obtains exactly the model above with $\tau = L / v$.  Conversely, the
measured $\tau$ can be converted into a coherence length $L = v\,\tau$ given
an independent wind-speed measurement.

### Fitting procedure

The fit is performed with `scipy.optimize.curve_fit` using the binned
correlation values as observations and the corresponding standard errors
($\sigma_C$ from the pair scatter in each bin) as weights (`sigma`,
`absolute_sigma=True`).  Only bins with $C > 0$ are included, since the
exponential model is positive-definite.  The parameter $\tau$ is bounded to
the physical interval $[0.1, 600]\,\mathrm{min}$ to prevent the optimiser
from wandering outside the lag window.  The fit uncertainties $\sigma_A$ and
$\sigma_\tau$ are extracted from the square root of the diagonal of the
covariance matrix returned by `curve_fit`.

The fit complements the empirical $1/e$ crossing read directly from the
binned $C(\Delta t)$: while the latter depends on the bin grid and the noise
level in the first crossing bin, the fit integrates information from all
positive-correlation bins and provides a more robust, reproducible estimate
with a propagated uncertainty suitable for publication.

```python
from scipy.optimize import curve_fit


def exp_decay(dt, A, tau):
    return A * np.exp(-dt / tau)


fit_mask = (good
            & (lag_min <= LAG_MAX_MIN)
            & (corr > 0)
            & np.isfinite(corr)
            & np.isfinite(corr_err)
            & (corr_err > 0))

x_fit = lag_min[fit_mask]
y_fit = corr[fit_mask]
e_fit = corr_err[fit_mask]

tau_guess = tau_min if np.isfinite(tau_min) else 60.0

try:
    popt, pcov = curve_fit(
        exp_decay, x_fit, y_fit,
        p0=[1.0, tau_guess],
        sigma=e_fit, absolute_sigma=True,
        bounds=([0, 0.1], [2.0, LAG_MAX_MIN]),
        maxfev=20_000
    )
    perr = np.sqrt(np.diag(pcov))
    A_fit, tau_fit_min = popt
    A_err, tau_fit_err = perr
    tau_fit_hr = tau_fit_min / 60.0
    print("=== Exponential fit  C(dt) = A * exp(-dt/tau) ===")
    print(f"  A   = {A_fit:.4f} +/- {A_err:.4f}")
    print(f"  tau = {tau_fit_min:.1f} +/- {tau_fit_err:.1f} min")
    print(f"      = {tau_fit_hr:.2f} +/- {tau_fit_err/60:.2f} h")
    fit_ok = True
except Exception as exc:
    print(f"Fit failed: {exc}")
    fit_ok = False
```

```python
fig5, axes5 = plt.subplots(1, 2, figsize=(16, 6))
fig5.subplots_adjust(wspace=0.32)

if fit_ok:
    t_dense_min = np.logspace(
        np.log10(LAG_MIN_MIN), np.log10(LAG_MAX_MIN), 500
    )
    t_dense_hr = t_dense_min / 60.0
    c_dense    = exp_decay(t_dense_min, *popt)

for ax5, xdata, x_dense, xlabel, xscale, xlim, x_tau in [
    (axes5[0], lag_hr,  t_dense_hr  if fit_ok else None,
     r"$\Delta t$ [hours]",   "linear",
     (0, LAG_MAX_MIN / 60.0),
     tau_fit_hr  if fit_ok else np.nan),
    (axes5[1], lag_min, t_dense_min if fit_ok else None,
     r"$\Delta t$ [minutes]", "log",
     (LAG_MIN_MIN, LAG_MAX_MIN),
     tau_fit_min if fit_ok else np.nan),
]:
    ax5.fill_between(
        xdata[good], corr[good] - corr_err[good], corr[good] + corr_err[good],
        alpha=0.25, color=CORR_COLOR
    )
    ax5.plot(xdata[good], corr[good],
             color=CORR_COLOR, lw=1.5, marker="o", ms=3.5, zorder=3,
             label=r"$C(\Delta t)$")
    if fit_ok:
        ax5.plot(x_dense, c_dense,
                 color="darkred", lw=2.0, ls="-", zorder=5,
                 label=(rf"$A\,e^{{-\Delta t/\tau}}$"
                        rf", $\tau={tau_fit_hr:.1f}$ h"))
        unit = xlabel.split('[')[1].rstrip(']') if '[' in xlabel else ''
        ax5.axvline(x_tau, ls="--", color=TAU_COLOR, lw=1.5,
                    label=rf"$\tau = {x_tau:.1f}$ {unit}")
    ax5.axhline(0.0,      ls="--", color=ZERO_COLOR,  lw=1.0)
    ax5.axhline(1.0/np.e, ls=":",  color=INV_E_COLOR, lw=1.5,
                label=r"$C = 1/e$")
    ax5.set_xscale(xscale)
    ax5.set_xlim(xlim)
    ax5.set_ylim(-1.5, 1.5)
    ax5.set_xlabel(xlabel, fontsize=13)
    ax5.set_ylabel(r"$C(\Delta t)$", fontsize=13)
    ax5.legend(fontsize=9, loc="upper right")

axes5[0].set_title("Exponential fit — linear hours", fontsize=11)
axes5[1].set_title("Exponential fit — log minutes",  fontsize=11)
fig5.suptitle(
    f"PWV correlation + exponential fit (0–10 h) — {version_run} (tight cuts)",
    fontsize=12, y=1.01
)
figfile5 = f"{pathfigs}/{prefix}_{version_run}_2point_corr_expfit{figtype}"
fig5.savefig(figfile5, bbox_inches="tight")
print(f"Saved: {figfile5}")
plt.show()
```
