## Configuration

```python
tag = "inst1_2d_asm_Nx_M2I1NXASM"
years = "2021-2025"
```

```python
input_path = "./Merradata"
```

```python
output_file = f"Merge_{tag}-{years}.csv"
```

## Select files

```python
list_of_files = os.listdir(input_path)
```

```python
selected_files = []
for file in list_of_files:
    if re.search(tag,file):
        selected_files.append(file)
```

```python
selected_files  = sorted(selected_files) 
selected_files
```

```python
NF = len(selected_files)
```

### Read the file

```python
all_df = []
for idx in range(NF):
    fullfilename = os.path.join(input_path,selected_files[idx])
    df = pd.read_csv(fullfilename)
    all_df.append(df)
```

```python
df = pd.concat(all_df)
```

```python
df
```

```python
# convert timestring into pandas datetime
df["Time"] = pd.to_datetime(df["time"])
```

```python
list(df.columns)
```

## Plot

```python
fig,ax = plt.subplots(1,1,figsize=(18,6),layout='constrained')
df.plot(x="Time",y="TQV",ax=ax,color= 'b',grid=True,rot=45)
from matplotlib.dates import DateFormatter
date_form = DateFormatter("%y-%m-%dT%H:%M")
ax.set_xlabel("time")
ax.xaxis.set_major_formatter(date_form)
#plt.xticks(rotation=30)
ax.set_title("MERRA2 : Precipitable water vapor")
ax.set_ylabel("PWV (mm)")
figfilename = f"M2-PWV-{years}{figtype}"
figfile = os.path.join(pathfigs,figfilename)
plt.savefig(figfile)
plt.show()
```

```python
fig,ax = plt.subplots(1,1,figsize=(18,6),layout='constrained')
df.plot(x="Time",y="TO3",ax=ax,color= 'b',grid=True,rot=45)
from matplotlib.dates import DateFormatter
date_form = DateFormatter("%y-%m-%dT%H:%M")
ax.set_xlabel("time")
ax.xaxis.set_major_formatter(date_form)
#plt.xticks(rotation=30)
ax.set_title("MERRA2 : Ozone")
ax.set_ylabel("Ozone (DU)")
figfilename = f"M2-Ozone-{years}{figtype}"
figfile = os.path.join(pathfigs,figfilename)
plt.savefig(figfile)
plt.show()
```

## Save merged file
