```python
selected_run1 = "run_v6"
selected_run2 = "run_v5"
```

```python
atmfilename1 = mergedextractedfilesdict [selected_run1]
tag1 = legendtag[selected_run1] 

atmfilename2 = mergedextractedfilesdict [selected_run2]
tag2 = legendtag[selected_run2] 
```

```python
print(f"Spectractor Extracted atmospheric parameters file 1 : {atmfilename1}  for config {tag1}")
print(f"Spectractor Extracted atmospheric parameters file 2 : {atmfilename2}  for config {tag2}")
```

```python
tag1_clean = re.sub(r'[^a-zA-Z0-9]+', '_', tag1)
tag2_clean = re.sub(r'[^a-zA-Z0-9]+', '_', tag2)
```

```python
print(tag1_clean)
print(tag2_clean)
```

```python
input_path1 = os.path.dirname(atmfilename1)
input_path2 = os.path.dirname(atmfilename2)
inputfilename1 = os.path.basename(atmfilename1)
inputfilename2 = os.path.basename(atmfilename2)

output_path = "data/spectro_joined"
outputfilename = f"auxtel_run_{tag1_clean}-join-{tag2_clean}.npy" 
fulloutputfilename = os.path.join(output_path ,outputfilename) 
```

```python
specdata1 = np.load(atmfilename1,allow_pickle=True)
specdata2 = np.load(atmfilename2,allow_pickle=True)
```

```python
df_spec1 = pd.DataFrame(specdata1)
df_spec2 = pd.DataFrame(specdata2)
```

```python
# add time for plotting
df_spec1["Time"] = pd.to_datetime(df_spec1["DATE-OBS"])
df_spec2["Time"] = pd.to_datetime(df_spec2["DATE-OBS"])
```

```python
df_spec1["nightObs"] = df_spec1.apply(lambda x: x['id']//100_000, axis=1)
df_spec2["nightObs"] = df_spec2.apply(lambda x: x['id']//100_000, axis=1)
```

```python
print((df_spec1["nightObs"].unique()))
```

```python
print((df_spec2["nightObs"].unique()))
```

```python
df_spec1 = df_spec1[df_spec1["nightObs"]<20250000]
```

```python
print(df_spec1["nightObs"].unique())
```

```python
df_joined = pd.concat([df_spec1 ,df_spec2])
```

```python
print(df_joined["nightObs"].unique())
```

```python
rec_array = df_joined.to_records(index=False)
```

```python
np.save(fulloutputfilename , rec_array)
```
