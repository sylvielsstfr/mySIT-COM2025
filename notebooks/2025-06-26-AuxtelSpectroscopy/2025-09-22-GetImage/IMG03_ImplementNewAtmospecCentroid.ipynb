## Extract the postISRCCD from the butler and find the peak positions

```python
NMAX = 30
```

```python
len(df)
```

```python
df
```

```python
# postisrccd container
all_postisrccds = []
# title of the figure
all_titles = []
# meta data associated to the exposure
all_md = []
# sequence number 
all_seqnum = []
# a dictionnary containing many info
ccd_dict = {}
#tasks results
task_qfm_results = []
task_peek_results = []

for idx, row in df.iterrows():
    print(idx)
    if idx< NMAX:
        the_exposure = row["id"]
        print(idx,the_exposure)
        the_num = row["seq_num"]
       
        the_filter = row["FILTER"]
        the_title = f"{the_num} ({the_filter})"
        the_postisrccd = butler.get('postISRCCD', exposure=the_exposure, detector=0, collections=collection, instrument='LATISS')
        #raw_img= butler.get('raw', dataId={'exposure': the_exposure, 'instrument': 'LATISS', 'detector': 0}, collections = ['LATISS/calib','LATISS/raw/all',] )
        #meta = raw_img.getMetadata()
        meta = the_postisrccd.getMetadata()
        md = meta.toDict()
        all_titles.append(the_title)
        all_postisrccds.append(the_postisrccd )
        all_md.append(md)
        all_seqnum.append(the_num)
        ccd_dict[the_num] = {"exposure": the_exposure, "filter":the_filter,"ccd": the_postisrccd,"md":md} 

        # process the peak finding by the tasks

        # process the  QuickFrameMeasurementTask
        qfm_result = qfm.run(the_postisrccd)
        qfm_centroid = qfm_result.brightestObjCentroid
        task_qfm_results.append(qfm_centroid)
        print("task_qfm_centroid :: ", qfm_centroid)

         # process the PeekExposureTask
        peek_result = peek.run(the_postisrccd)
        peek_centroid = peek_result.brightestCentroid 
        # do not use all the info in the table for the moment
        #many_peeks_array = peek_result.table[["centroid_x", "centroid_y"]].as_array()
        task_peek_results.append(peek_centroid)
        print("task_peek_centroid :: ", peek_centroid)
        
    else:
        break
```

## Plots

```python
afwDisplay.setDefaultBackend("firefly")
```

```python
N = len(all_postisrccds)
for count in range(N):
    display = afwDisplay.Display(frame=count + 1)
    # cannot succeed to show white stars on dark sky
    display.setImageColormap("gray")
    display.scale("asinh", "zscale")
    display.mtv(all_postisrccds[count].image, title=all_titles[count])    

    with display.Buffering():

        xpix = task_qfm_results[count][0]
        ypix = task_qfm_results[count][1]
        display.dot("o", xpix, ypix, size=100, ctype="red")

        
        the_task_peek_result = task_peek_results[count]
        xpix = the_task_peek_result[0]
        ypix = the_task_peek_result[1]
        display.dot("+", xpix, ypix, size=100, ctype="blue")
```
